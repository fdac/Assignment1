```python
from __future__ import division
import nltk
from nltk import *
import re

#Open file and read
f = open('P&P.txt', 'r')
text = f.read()

f2 = open('HF.txt', 'r')
text2 = f2.read()

#Find the number of words in the text
num_words_book1= len(text)
print "Number of words_JaneAusten_Pride&Prejudice = " + str(num_words_book1)

num_words_book2= len(text2)
print "Number of words_MarkTwain_HuckleBerryFinn = " + str(num_words_book2)		

word1 = text.count("would")
print "would_count = " + str(word1)

probability1 = word1/num_words_book1
print "p_would_book1 = "+ str(probability1)

word1 = text2.count("would")
print "would_count = " + str(word1)

probability2 = word1/num_words_book2
print "p_would_book2 = "+ str(probability2)


word2 = text.count("must")
print "must_count = " + str(word2)

probability1 = word2/num_words_book1
print "p_must_book1 = "+ str(probability1)

word2 = text2.count("must")
print "must_count = " + str(word2)

probability2 = word2/num_words_book2
print "p_must_book2 = "+ str(probability2)


word3 = text.count("miss")
print "miss_count = " + str(word3)

probability1 = word3/num_words_book1
print "p_miss_book1 = "+ str(probability1)

word3 = text2.count("miss")
print "miss_count = " + str(word3)

probability2 = word3/num_words_book2
print "p_miss_book2 = "+ str(probability2)


word4 = text.count("agreeable")
print "agreeable_count = " + str(word4)

probability1 = word4/num_words_book1
print "p_agreeable_book1 = "+ str(probability1)

word4 = text2.count("agreeable")
print "agreeable_count = " + str(word4)

probability2 = word4/num_words_book2
print "p_agreeable_book2 = "+ str(probability2)


word5 = text.count("dear")
print "dear_count = " + str(word5)

probability1 = word5/num_words_book1
print "p_dear_book1 = "+ str(probability1)

word5 = text2.count("dear")
print "dear_count = " + str(word5)

probability2 = word5/num_words_book2
print "p_dear_book2 = "+ str(probability2)


word6 = text.count("scruple")
print "scruple_count = " + str(word6)

probability1 = word6/num_words_book1
print "p_scruple_book1 = "+ str(probability1)

word6 = text2.count("scruple")
print "scruple_count = " + str(word6)

probability2 = word6/num_words_book2
print "p_scruple_book2 = "+ str(probability2)
```
