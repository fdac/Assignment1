### Example 1. Dictionaries in python (associative arrays)

Plot the frequency distribution of words on a web page.

````python
import requests, re
# re is a module for regular expressions: to detect various combinations of characters
import operator

# Start from a simple document
r = requests .get('http://eecs.utk.edu')

# What comes back includes headers and other HTTP stuff, get just the body of the response
t = r.text

# obtain words by splitting a string using as separator one or more (+) space/like characters (\s) 
wds = re.split('\s+',t)

# now populate a dictionary (wf)
wf = {}
for w in wds:
    if w in wf: wf [w] = wf [w] + 1
    else:  wf[w] = 1

# dictionaries can not be sorted, so lets get a sorted *list*        
wfs = sorted (wf .iteritems(), key = operator .itemgetter (1), reverse=True)   

# lets just have no more than 15 words 
ml = min(len(wfs),15)
for i in range(1,ml,1):
    print wfs[i][0]+"\t"+str(wfs[i][1])   
````

use BeautifulSoup and nltk modules and practice some regular expressions.

````python
import requests, re, nltk
from BeautifulSoup import BeautifulStoneSoup
from nltk import clean_html
from collections import Counter
import operator

# we may not care about the usage of stop words
nltk .download ('stopwords')
#stop_words = nltk.corpus.stopwords.words('english') + [
# 'ut', '\'re','.', ',', '--', '\'s', '?', ')', '(', ':', '\'',
# '\"', '-', '}', '{', '&', '|', u'\u2014' ]

#enhanced by rjlusk10
stop_words = nltk.corpus.stopwords.words('english') + [
 'ut', '\'re','.', ',', '--', '\'s', '?', ')', '(', ':', '\'', 
 '\"', '-', '}', '{', '&', '|', u'\u2014', 'gutenberg', 'gutenberg-tm', 'electronic', 'terms',
'foundation', 'agreement', 'license', 'full', 'work', 'works', 'donations','must']


# We most likely would like to remove html markup
def cleanHtml (html):
    return BeautifulStoneSoup(clean_html(html),
                convertEntities=BeautifulStoneSoup.HTML_ENTITIES).contents[0]

# We also want to remove special characters, quotes, etc. from each word
def cleanWord (w):
    # r in r'[.,"\']' tells to treat \ as a regular character 
    # but we need to escape ' with \'
    # any character between the brackets [] is to be removed 
    return re.sub('[,"\.\'&\|:]', "", w)
       
# define a function to get text/clean/calculate frequency
def get_wf (URL):
    # first get the web page
    r = requests .get(URL)
    
    # Now clean
    # remove html markup
    t = cleanHtml (r .text) .lower()
    
    # split string into an array of words using any sequence of spaces "\s+" 
    wds = re .split('\s+',t)
    
    # remove periods, commas, etc stuck to the edges of words
    for i in range(len(wds)):
        wds [i] = cleanWord (wds [i])
    
    # If satisfied with results, lets go to the next step: calculate frequencies
    # We can write a loop to create a dictionary, but 
    # there is a special function for everything in python
    # in particular for counting frequencies (like function table() in R)
    wf = Counter (wds)
    
    # Remove stop words from the dictionary wf
    for k in stop_words:
        wf. pop(k, None)
        
    #how many regular words in the document?
    tw = 0
    for w in wf:
       tw += wf[w] 
        
    
    # Get ordered list
    wfs = sorted (wf .iteritems(), key = operator.itemgetter(1), reverse=True)
    ml = min(len(wfs),15)

    #Reverse the list because barh plots items from the bottom
    return (wfs [ 1:ml ] [::-1], tw)
        
# Now populate two lists    
(wf_ee, tw_ee) = get_wf('http://eecs.utk.edu')
(wf_bu, tw_bu) = get_wf('http://bus.utk.edu')
````

### Extra Credit:

The word I will analyze is the word "one." See the code below for the computation of binomial distributions for this word.

````python
import decimal

def get_wf1 (URL, word):
    # first get the web page
    r = requests .get(URL)
    
    # Now clean
    # remove html markup
    t = cleanHtml (r .text) .lower()
    
    # split string into an array of words using any sequence of spaces "\s+" 
    wds = re .split('\s+',t)
    
    # remove periods, commas, etc stuck to the edges of words
    for i in range(len(wds)):
        wds [i] = cleanWord (wds [i])
    
    # If satisfied with results, lets go to the next step: calculate frequencies
    # We can write a loop to create a dictionary, but 
    # there is a special function for everything in python
    # in particular for counting frequencies (like function table() in R)
    wf = Counter (wds)
    
    # Remove stop words from the dictionary wf
    for k in stop_words:
        wf. pop(k, None)
        
    #how many regular words in the document?
    tw = 0
    for w in wf:
       tw += wf[w] 

    #Return the count of the given word and the total words
    decimal.getcontext().prec = 6
    p = decimal.Decimal(wf[word]) / decimal.Decimal(tw)
    n = decimal.Decimal(tw)
    return (p, n)

def binom_95_confint(URL, word, title):
    p, n = get_wf1(URL, word)
    high = p + (decimal.Decimal(1.96) * (p / n * (1 - p)) ** decimal.Decimal(0.5))
    low = p - (decimal.Decimal(1.96) * (p / n * (1 - p)) ** decimal.Decimal(0.5))
    print "%s: P: %f, LB: %f, UB: %f" % (title, p, low, high)

       
# Populate four lists 
word = "one"
binom_95_confint('http://www.gutenberg.org/files/8799/8799-h/8799-h.htm', word, "Paradise")
binom_95_confint('http://www.gutenberg.org/files/8789/8789-h/8789-h.htm', word, "Hell")
binom_95_confint('http://www.gutenberg.org/files/2701/2701-h/2701-h.htm', word, "Moby Dick")
binom_95_confint('http://www.gutenberg.org/files/76/76-h/76-h.htm', word, "Huck Finn")
````

To perform this analysis, I modified the get_wf function from earlier and had it return instead the p and n values to be used in the binomial function. I then created a function binom_95_confint that calls get_wf and computes a 95% confidence interval based on the returned p and n values.

The output above displays confidence intervals for the four different works analyzed in parts 1 and 2. P is the proportion of the word "one" relative to total words, LB is the lower bound of the 95% confidence interval, and UB is the upper bound of the interval (using a normal approximation).

These results are exactly the opposite of what I expected. On the lists above showing "Paradise" and "Hell," the word "one" occupies the second spot. It would seem, therefore, that the proportional use of this word would be similar for both works. The confidence intervals, however, do not overlap.

Conversely, the word "one" is the most used word in "Moby Dick," and is only the seventh most used word in "Huck Finn." However, these confidence intervals overlap, indicating that the word is not necessarily more common to one author or the other.

````python
local = False #Whether or not to use local copies
num_compare = 15 #How many words to display/compare

#Function that returns a sorted list of the words in the page at 'URL'
def get_words(URL):
    
    if not local:
        #Get the website and clean HTML
        page = requests.get(URL)
        text = BeautifulStoneSoup(clean_html(page.text), convertEntities=BeautifulStoneSoup.HTML_ENTITIES).contents[0]
    else:
        #Local copies don't need to be retrieved and they are already in text format
        text = BeautifulStoneSoup(clean_html(URL), convertEntities=BeautifulStoneSoup.HTML_ENTITIES).contents[0]
    
    #Separate out the words in the text
    words = re.split('\s+',text.lower())
    
    #Remove punctuation from words
    for i in range(len(words)):
        words[i] = re.sub(r'[.,"\.]', "", words[i]) #This isn't perfect
    
    #Get the frequency of each word
    word_dict = Counter(words)
    
    #and remove stop words from it
    for word in stop_words:
        word_dict.pop(word, None)
    
    #Sort the list of words
    result = sorted(word_dict.iteritems(), key = operator.itemgetter(1), reverse = True)
    
    #Return the words in the list in reverse order (for numpy)
    return result

%matplotlib inline
def plot_results(words1, words2, title):
    
    #Reverse order
    words1 = words1[::-1]
    words2 = words2[::-1]
    
    #Find the number of common words
    common = 0
    wc1 = [w[0] for w in words1]
    wc2 = [w[0] for w in words2]
    for word in wc1:
        if word in wc2:
            common = common + 1
    
    title = title + "\n- " + str(common) + " common words -"
    
    fig = plt.figure(figsize = (10, 8))
    fig.suptitle(title, fontsize = 20)
    
    ax = fig.add_subplot(111)
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_color('none')
    ax.spines['left'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.tick_params(labelcolor = 'w', top = 'off', bottom = 'off', left = 'off', right = 'off', labelsize = 20)
    
    #Create the two plots
    plot1 = fig.add_subplot(121)
    plt.subplots_adjust(wspace = 0.5)
    
    pos = np.arange(len(words1) + 1) 
    plot1.tick_params(axis='both', which='major', labelsize=12)
    pylab.yticks(pos, [ x[0] for x in words1 ])
    plot1.barh(range(len(words1)), [ x[1] for x in words1 ], align='center')
    
    plot2 = fig.add_subplot(122)
    plt.subplots_adjust(wspace = 0.5)
    
    pos = np.arange(len(words2) + 1) 
    plot2.tick_params(axis='both', which='major', labelsize=12)
    pylab.yticks(pos, [ x[0] for x in words2 ])
    plot2.barh(range(len(words2)), [ x[1] for x in words2 ], align='center')

#The four books being used to compare authors:
#  - The Adventures of Tom Sawyer - Mark Twain
#  - Life on the Mississippi - Mark Twain
#  - The Jungle - Upton Sinclair
#  - Siddhartha - Hermann Hesse    

if not local:
    tom_sawyer = "http://www.gutenberg.org/files/74/74-h/74-h.htm"
    mississippi = "http://www.gutenberg.org/files/245/245-h/245-h.htm"
    jungle = "http://www.gutenberg.org/files/140/140-h/140-h.htm"
    siddhartha = "http://www.gutenberg.org/files/2500/2500-h/2500-h.htm"
else:
    #Option to use local copies so I don't lock myself out of Project Gutenberg (again)
    with open('../gutenberg/tom_sawyer.html', 'r') as ts:
        tom_sawyer = ts.read()
    with open('../gutenberg/mississippi.html', 'r') as m:
        mississippi = m.read()
    with open('../gutenberg/jungle.html', 'r') as j:
        jungle = j.read()
    with open('../gutenberg/siddhartha.html', 'r') as s:
        siddhartha = s.read()

#Grab the full list of words from each book
ts_words = get_words(tom_sawyer)
m_words = get_words(mississippi)
j_words = get_words(jungle)
s_words = get_words(siddhartha)

#Get the total number of words overall from each book
ts_total = sum([w[1] for w in ts_words])
m_total = sum([w[1] for w in m_words])
j_total = sum([w[1] for w in j_words])
s_total = sum([w[1] for w in s_words])

#Find each word's binomial distribution
ts_bdist = [ (w[0], w[1]/float(ts_total)) for w in ts_words ]
m_bdist = [ (w[0], w[1]/float(m_total)) for w in m_words ]
j_bdist = [ (w[0], w[1]/float(j_total)) for w in j_words ]
s_bdist = [ (w[0], w[1]/float(s_total)) for w in s_words ]

#The above can be limited to [:num_compare] to reduce computation time

title1 = "Comparing the top " + str(num_compare) + " words in \'The Adventures of Tom Sawyer\' and \'Life on the Mississippi\'"
plot_results( ts_words[:num_compare], m_words[:num_compare], title1 )

title2 = "Comparing the top " + str(num_compare) + " words in \'The Jungle\' and \'Siddhartha\'"
plot_results( j_words[:num_compare], s_words[:num_compare], title2 )


print "Word\t\tBinomial Dist."
for word in ts_bdist[:2]:
    print word[0] + ":\t\t" + str(word[1])
print "--------------------"
for word in m_bdist[:2]:
    print word[0] + ":\t\t" + str(word[1])
print "--------------------"
for word in j_bdist[:2]:
    print word[0] + ":\t\t" + str(word[1])
print "--------------------"
for word in s_bdist[:2]:
    print word[0] + ":\t\t" + str(word[1])
print "--------------------"
````

````python
from math import sqrt
# comparing distribution for the word "man" in Dostoyevsky's novels
print "house of death word count: %s \nnotes from the underground word count: %s\n" % (wc_hod, wc_notes)
print "proportion of all words that are \"man\" in \"House of Death\": " + str( float(wf_hod[11][1])/wc_hod )
print "proportion of all words that are \"man\" in \"Notes From the Underground\": " + str( float(wf_notes[9][1])/wc_notes ) \
    + "\n"

# differenct of proportions
pManHod = float(wf_hod[11][1])/wc_hod
pManNotes = float(wf_notes[9][1])/wc_notes

diff = pManHod - pManNotes

SEhod = pManHod * (1 - pManHod) / wc_hod
SEnotes = pManNotes * (1 - pManNotes) / wc_notes
SEtest = sqrt( SEhod + SEnotes )

LCI = diff - SEtest # lower bound of confidence interval (CI)
UCI = diff + SEtest # upper bound of CI

print  "Difference between the proportions = " + str(round(diff, 4))
print  "Lower bound 95% confidence interval = " + str(round(LCI, 4))
print  "Upper bound 95% confidence interval = " + str(round(UCI, 4))
````

The 95% confidence interval for the difference between the proportions of the word "man" in Dostoyevsky's "House of the Dead" and "Notes From the Underground" contains 0. This is evidence that there is no statistically significant difference between the relative frequency of occurence for this word.

````python
# comparing distribution for the word "little" in "Alice's Adventures in Wonderland" and "The Adventures of Sherlock Holmes"
print "Alice's Adventures in Wonderland word count: " + str(wc_alice)
print "The Adventures of Sherlock Holmes word count: " + str(wc_sherlock) + "\n"
print "proportion of all words that are \"little\" in \"Alice's Adventures in Wonderland\": " \
    + str( float(wf_alice[1][1])/wc_alice )
print "proportion of all words that are \"little\" in \"The Adventures of Sherlock Holmes\": " \
    + str( float(wf_sherlock[6][1])/wc_sherlock ) + "\n"

# differenct of proportions
pA = float(wf_alice[1][1])/wc_sherlock
pS = float(wf_sherlock[9][1])/wc_sherlock

diff = pA - pS

SEalice = pA * (1 - pA) / wc_alice
SEsherlock = pS * (1 - pS) / wc_sherlock
SEtest = sqrt( SEalice + SEsherlock )

LCI = diff - SEtest # lower bound of confidence interval (CI)
UCI = diff + SEtest # upper bound of CI

print  "Difference between the proportions = " + str(round(diff, 4))
print  "Lower bound 95% confidence interval = " + str(round(LCI, 4))
print  "Upper bound 95% confidence interval = " + str(round(UCI, 4))
````

The 95% confidence interval for the difference between the proportions of the word "little" in "Alice's Adventures in Wonderland" and "The Adventures of Sherlock Holmes" does not contain 0. Both the upper and lower bound of the confidence interval are negative, offering evidence that "little" has a statistically significantly higher probability of occurrence in Arthur Conan Doyal's books than they do in Lewis Carroll's.

````python
def get_wf_k (URL, writer):
    # first get the web page
    r = requests .get(URL)
    
    # Now clean
    # remove html markup
    t = cleanHtml (r .text) .lower()
    
    # split string into an array of words using any sequence of spaces "\s+" 
    wds = re .split('\s+',t)
    
    # remove periods, commas, etc stuck to the edges of words
    for i in range(len(wds)):
        wds [i] = cleanWord (wds [i])
    
    # If satisfied with results, lets go to the next step: calculate frequencies
    # We can write a loop to create a dictionary, but 
    # there is a special function for everything in python
    # in particular for counting frequencies (like function table() in R)
    wf = Counter (wds)
    
    # Remove stop words from the dictionary wf
    for k in stop_words:
        wf. pop(k, None)
        
    #how many regular words in the document?
    tw = 0
    for w in wf:
       tw += wf[w] 
            
    # Get ordered list
    wfs = sorted (wf .iteritems(), key = operator.itemgetter(1), reverse=True)
    ml = min(len(wfs),15)
    wl = max(len(wfs),20)
    
    print "\n\nWord" + "\t" + "Frequency" + "\t" + "Probability"
    for i in range(1,wl,1):
        print wfs[i][0]+"\t"+str(wfs[i][1]) + "\t" + str(float(wfs[i][1])/float(tw))    
        if (i > 5):
            break
    
    if (writer == 'mt'):
        print "\nFor most used word depending on writing style of Mark Twain"    
        print " Word" + "\t" + "Frequency" + "\t" + "Probability"
        for i in range(1,wl,1):
            if (wfs[i][0] == 'would' or wfs[i][0] == 'must' or wfs[i][0] == 'miss' or wfs[i][0] == 'agreeable'  or wfs[i][0] == 'terrific'):
                print wfs[i][0]+"\t"+str(wfs[i][1]) + "\t" + str(float(wfs[i][1])/float(tw))

    if (writer == 'ja'):
        print "\nFor most used word depending on the writing style of Jane Austin"    
        print " Word" + "\t" + "Frequency" + "\t" + "Probability"
        for i in range(1,wl,1):
            if (wfs[i][0] == 'agreeable' or wfs[i][0] == 'dear' or wfs[i][0] == 'irrepressible' or wfs[i][0] == 'scruple'  or wfs[i][0] == 'terrific'):
                print wfs[i][0]+"\t"+str(wfs[i][1]) + "\t" + str(float(wfs[i][1])/float(tw))

    #Reverse the list because barh plots items from the bottom
    return (wfs [ 1:ml ] [::-1], tw)
        
# Now populate two lists    
(wf_ee, tw_ee) = get_wf_k('http://www.gutenberg.org/ebooks/1342.txt.utf-8', 'ja') # Jane Austin
(wf_bu, tw_bu) = get_wf_k('http://www.gutenberg.org/ebooks/76.txt.utf-8', 'mt') # Mark Twain
````
